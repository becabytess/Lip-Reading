```python
import os 
import numpy as np
data_dir = os.path.join("raw_dataset","1","data")
speakers = os.listdir(data_dir)

```

```python
import tarfile 
from tqdm import tqdm
os.makedirs("compressed_data", exist_ok=True)
for speaker in tqdm(speakers):
    speaker_dir = os.path.join(data_dir, speaker)
    tar_path = os.path.join("compressed_data", f"{speaker}.tar.gz")
    with tarfile.open(tar_path, "w:gz") as tar:
        tar.add(speaker_dir, arcname=speaker)   
    tar.close()
```

```python

from huggingface_hub import create_repo
import huggingface_hub as hf
import os
from dotenv import load_dotenv
from huggingface_hub import login



load_dotenv()
hf_token = os.getenv("hf_token")

create_repo(
        repo_id="bekalemu/grid-corpus-compressed",
        private=False  ,
        token=hf_token,
        repo_type="dataset",
        exist_ok=True
    )



repo_id = "bekalemu/grid-corpus-compressed"

#push the dataset to repo

data_dir = "compressed_data"

hf.upload_folder(
    repo_id=repo_id,
    folder_path=data_dir,
    token=hf_token,
    repo_type="dataset",
    commit_message="Initial commit of the dataset",
    commit_description="This commit contains the initial version of the dataset.",
)


```
